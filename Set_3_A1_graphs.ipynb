```python
import matplotlib.pyplot as plt
import numpy as np


!git clone https://github.com/Stars007z/Set_3_A1.git

file_path = "Set_3_A1/A1_additional/results.txt"

n_values = []
areas = []
deviations = []

f = False
with open(file_path, "r") as file:
    for line in file:
        if not f:
            f = True
            continue
        n, area, deviation = map(float, line.strip().split())
        n_values.append(int(n))
        areas.append(area)
        deviations.append(deviation)


plt.figure(figsize=(10, 5))
plt.plot(n_values, areas, label="Approximated Area", color="blue")
plt.axhline(y=0.944517, color="red", linestyle="--", label="Exact Area")
plt.title("Approximated Area vs N")
plt.xlabel("Number of Points (N)")
plt.ylabel("Approximated Area")
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(n_values, deviations, label="Relative Deviation", color="green")
plt.title("Relative Deviation vs N")
plt.xlabel("Number of Points (N)")
plt.ylabel("Relative Deviation")
plt.legend()
plt.grid()
plt.show()
```
